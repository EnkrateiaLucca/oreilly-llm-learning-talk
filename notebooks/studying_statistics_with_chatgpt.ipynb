```python
# Probability of having the disease (P(Disease))
prob_disease = 0.01

# Probability of testing positive given the disease (P(Pos|Disease))
prob_pos_given_disease = 0.9

# Probability of not having the disease (P(No Disease))
prob_no_disease = 1 - prob_disease

# Probability of testing positive given no disease (P(Pos|No Disease))
prob_pos_given_no_disease = 0.1

# Bayes' Theorem to find P(Disease|Pos)
prob_disease_given_pos = (prob_pos_given_disease * prob_disease) / ((prob_pos_given_disease * prob_disease) + (prob_pos_given_no_disease * prob_no_disease))

print(f"The probability of having the disease given a positive test is {prob_disease_given_pos}")
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Generating a normal distribution data
mean = 0
std_dev = 1
samples = np.random.normal(mean, std_dev, 1000)

# Visualizing the distribution
sns.histplot(samples, kde=True)
plt.title('Normal Distribution (mean=0, std_dev=1)')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

```python
# Generating binomial distribution data
n = 10  # number of trials
p = 0.5  # probability of success
binomial_data = np.random.binomial(n, p, 1000)

# Visualizing the distribution
sns.histplot(binomial_data, bins=10, kde=False)
plt.title('Binomial Distribution (n=10, p=0.5)')
plt.xlabel('Number of Successes')
plt.ylabel('Frequency')
plt.show()
```
